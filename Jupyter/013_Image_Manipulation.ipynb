```python
#
# Image Manipulation Technique
#
# @author becxer
# @email becxer87@gmail.com
# @reference https://github.com/sjchoi86/Tensorflow-101
#

import os, sys
from scipy.misc import imread, imresize
import skimage.io
import skimage.transform

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

print ("packages are loaded")
```

```python
print ("test images : " + str(os.listdir('testimg')))
```

```python
# read image and show
dva = imread('testimg/dva.jpg')

plt.figure(0)
plt.imshow(dva)
plt.title('original dva online! ' + str(dva.shape))
plt.draw()
```

```python
# reverse rgb
dva_float = imread('testimg/dva.jpg').astype(np.float)

plt.figure(0)
plt.imshow(dva_float)
plt.title('reversed dva online! ' + str(dva_float.shape))
plt.draw()
```

```python
# reverse reverse rgb
dva = imread('testimg/dva.jpg').astype(np.float)

plt.figure(0)
plt.imshow(dva/255)
plt.title('reverse reversed dva online! ' + str(dva.shape))
plt.draw()
```

```python
#resize image
small_dva = imresize(dva, [100,100,3])

plt.figure(0)
plt.imshow(small_dva)
plt.title('small dva online! ' + str(small_dva.shape))
plt.draw()
```

```python
#change to grayscale
def rgb2gray(rgb_img):
    if len(rgb_img.shape) == 3:
        return np.dot(rgb_img,[0.299,0.587,0.114])
    else:
        return rgb_img

gray_dva = rgb2gray(small_dva)

plt.figure(0)
plt.imshow(gray_dva, cmap=plt.get_cmap("gray"))
plt.title('gray dva online! ' + str(gray_dva.shape))
plt.draw()
```

```python
#change to vector
dva_row = np.reshape(gray_dva, (1,-1))
print ("dva_row shape : " + str(dva_row.shape))

#change to tensor
dva_tensor= np.reshape(dva_row, (100,100))
print ("dva_row shape : " + str(dva_tensor.shape))
```
